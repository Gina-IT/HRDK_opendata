```python
import pandas as pd
```

## 파일데이터 리스트

```python
file_data= pd.read_csv("./data.go/파일시스템.csv", encoding='cp949')
file_data
```

```python
# null값 채우기 : fillna("채우고 싶은 문자 데이터") or fillna(채우고 싶은 숫자 값)

file_data= file_data.fillna(" ")
file_data.tail(10)
```

```python
# iloc를 사용할 때 ":" 앞쪽 인덱스의 데이터는 포함하지만 뒤쪽 인덱스의 데이터는 포함하지 않음
# 따라서  ":" 뒤쪽에는 포함하고 싶은 데이터까지의 인덱스 +1의 인덱스를 적어주어야 함

file_data.iloc[0:3,:]
```

```python
# 데이터 타입 확인 : dtyoes

file_data.dtypes
```

### 파일데이터 리스트 부서별 분할

정보화사업 1부 띄어쓰기로 데이터 한 행이 순서가 맞지 않아서 분할 후 같은 사업부 데이터 프레임 병합하고 인덱스 조작하여 재정렬함

```python
start=0
end=1
file_dfs=[]

while end < len(file_data):
    if file_data.iloc[end,1] != file_data.iloc[end-1,1]:
        df=file_data.iloc[start:end,:]
        file_dfs.append(df)
        start = end
    end = end+1
if end== len(file_data):
    df=file_data.iloc[start:end, :]
    file_dfs.append(df)
```

```python
# list 길이 확인 : len(list)

len(file_dfs)
```

```python
file_team1= file_dfs[0]
file_team2= file_dfs[1]
file_team3= file_dfs[2]
file_team4= file_dfs[3]
file_team5= file_dfs[4]
file_team6= file_dfs[5]
```

```python
# 데이터프레임 병합 : concat([df1. df2])- 그냥 물리적으로 이어붙임  (merge - 고유값을 기준으로 이어붙임)

file_team1= pd.concat([file_team1, file_team5])
file_team1
```

```python
file_team5= file_team6
file_team5
```

```python
file_team2=file_team2.reset_index(drop=True)
file_team3=file_team3.reset_index(drop=True)
file_team4=file_team4.reset_index(drop=True)
file_team5=file_team5.reset_index(drop=True)
```

```python
tmp= file_team1.reset_index()
tmp
```

```python
tmp.iloc[20,0]=4.5
tmp= tmp.sort_values('index')
tmp
```

```python
file_team1= tmp.drop('index',axis=1)
file_team1
```

```python
file_team1= file_team1.reset_index(drop=True)
file_team1
```

```python
# 부서별 제공하는 파일 데이터 데이터프레임 csv파일로 내보내기

file_team1.to_csv('file_team1.csv', index=False)
file_team2.to_csv('file_team2.csv', index=False)
file_team3.to_csv('file_team3.csv', index=False)
file_team4.to_csv('file_team4.csv', index=False)
file_team5.to_csv('file_team5.csv', index=False)
```
