### 반복문을 이용해 여러 csv파일 한번에 read하기
1. 'df_error_files'라는 이름의 딕셔너리 생성(dict{})
2. 반복문을 이용해 파일명을 저장한 리스트 'error_files'에 파일명으로 접근 (인덱스 번호로 접근 X)
4. 파일 번호를 key값으로 사용하기 위해, 파일명 앞 두글자를 읽어들여 'file_num' 변수에 담음
5. csv파일을 읽어오는데, encoding 옵션을 'utf-8'로 해보고, 안되면 'cp949'로 하여 파일을 읽어 'df'변수에 담음

### 반올림 함수 생성
round()함수의 경우 사사오입의 원칙이 적용되어, 반올림할 자리의 수가 홀수 이면 올림하지만 짝수일 경우 내림을 한다.  
따라서 사사오입의 원칙이 적용되지 않는 반올림을 구현하고 싶다면 함수를 직접생성해야 한다.
* 소수 둘째 자리에서 반올림하여 첫째 자리까지 표시할 것을 가정하고 만든 함수임을 유의
1. 원래 숫자에 10을 곱하여 소수점을 한칸 뒤로 이동시켜준다
2. 소수('multiply_num')에서 정수화 한 수('int(multiply_num)')을 빼주면 소수점 이하의 수만 남게 되는데, 이 때 소수점 이하의 수가 0.5 이상이면
3. 정수화 한 수에 1을 더해주고 더해준 수를 'roundup_num'이라는 변수에 저장한다
4. 0.5보다 작을 경우 그냥 정수화 한 수를 'roundup_num'이라는 변수에 저장한다
5. 함수 초반에 원래 수 'num'에 10을 곱해주었기 때문에, 원래의 소수점 위치로 이동시키기 위해 'roundup_num'을 다시 10으로 나누어준다
6. 계산 식을 'final_num'변수에 저장하여 return한다

```python
def roundUp(num):
    roundup= []
    if math.isnan(num):
        rounup_num=0
    else:
        multiply_num= num*10
        if (multiply_num - int(multiply_num)) >= 0.5: 
            roundup_num= int(multiply_num) + 1 
        else: 
            roundup_num= int(multiply_num)
        roundup.append(roundup_num/10)
        
    to_return= pd.Series(roundup)
    return to_return
```

### 각 csv파일의 오류 확인 및 오류가 있는 행 추출

```python
def find_error(file_num, which_file):
    # 한 파일 내 오류가 있는 행들을 저장할 딕셔너리 생성
    all_error_rows= {}
    
    print('파일 번호: ',file_num)
    
    check_examinees_num= (which_file['필기시험접수'] >= which_file['필기시험응시']) &  (which_file['필기시험응시'] >= which_file['필기시험합격']) & (which_file['실기시험접수'] >= which_file['실기시험응시']) &  (which_file['실기시험응시'] >= which_file['실기시험합격'])
    if False in check_examinees_num.unique():
        print('응시자수 오류 있음')
        # 응시자수에 오류가 있는 경우 'examinees_error_rows'를 key값으로 하여 딕셔너리에 추가
        all_error_rows['examinees_error_rows']=np.where(check_examinees_num ==False)[0]  
    else:
        print('응시자수 오류 없음')
     
    check_percentage= (which_file['필기시험합격률(%)'].iloc[0] == roundUp(which_file['필기시험합격'].iloc[0]/which_file['필기시험응시'].iloc[0]*100)) & (which_file['실기시험합격률(%)'].iloc[0] == roundUp(which_file['실기시험합격'].iloc[0]/which_file['실기시험응시'].iloc[0]*100))
    if False in check_percentage.unique():
        print('합격률 오류 있음')
        all_error_rows['percentage_error_rows']=np.where(check_percentage ==False)[0]
    else:
        print('합격률 오류 없음')
     
    # 응시자수, 합격률, 자격취득자 수 오류 행을들 저장한 딕셔너리를 하나의 데이터프레임으로 생성
    # 딕셔너리의 길이가 다르기 때문에 하나의 데이터프레임으로 만들기 위해 다음과 같이 작성
    if bool(all_error_rows):    
        return all_error_rows
```

```python
def find_license_error_rows(file_num, which_file):
    check_license= which_file['자격취득자현황(명)'] <= which_file['실기시험합격']
    if False in check_license.unique():
        print('자격취득자 수가 실기시험합격자 수보다 많은 종목이 있습니다')
        license_error_rows=np.where(check_license ==False)[0]
        
    print('\n')
    
    return license_error_rows
```

```python
# 각 csv파일 별로 오류가 있는 행을 정리한 데이터프레임을 파일 번호를 key값으로 하여 저장할 딕셔너리 생성
all_error_dfs= {}
license_error_lists={}

for key in df_error_files:    
    which_file= df_error_files[key].dropna()     # 계산 오류를 막기 위해 NaN값 미리 제거
    
    # NaN값을 제거 한 데이터 'which_file'과 key값을 함수 인자로 넣어줌
    # 함수의 return 값을 'errors_df'라는 변수로 받음
    errors_rows= find_error(key, which_file)    
    
    license_error_rows= find_license_error_rows(key, which_file)  
    license_error_lists[key]= list(license_error_rows)
    
    # 받은 return값을 'all_error_dfs' 딕셔너리에 추가
    # 이때 return받은 것은 해당 파일의 오류 행들을 모은 데이터 프레임
    # 각 파일의 파일 번호를 key값, 해당 데이터 프레임을 value로 하여 딕셔너리에 추가함
    if errors_rows is not None:
        all_error_dfs[key]= pd.DataFrame(errors_rows)
    
print(all_error_dfs.keys())
print(license_error_lists.keys())
```
